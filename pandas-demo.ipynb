### apply()

```python
# do operation on each cloumn of the df

def date_to_year(row):
    return int(row['date'].split('-')[0])

df.apply(date_to_year, axis=1) #axis = 1 for rows
```

```python
df['year'] = df.apply(date_to_year, axis=1)
```
